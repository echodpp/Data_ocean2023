## MLD Visualization 

```python
import pandas as pd
import numpy as np

df_mld = pd.read_csv('mld_means_new.csv')
df_mld.sample(5)
```

```python
month_name = ["01","02","03","04","05","06","07","08","09","10","11","12"]
for month in month_name:
    unique_values = df_mld['mld_' + month].unique()
    print(f"Unique values in mld_{month}: {unique_values}")
df_mld['yearly_mean_mld'] = df_mld[['mld_' + month for month in month_name]].mean(axis=1)
```

```python
df_mld.sample(5)
```

```python
for col in df_mld.columns:
    if 'mld' in col:
        print(f"Correlation between 'Cr_nmol/kg' and '{col}': {df_mld['Cr_nmol/kg'].corr(df_mld[col])}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

mld_cols = ['yearly_mean_mld']

for col in mld_cols:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df_mld, x=col, y='Cr_nmol/kg', hue='Ocean')
    plt.title(f'Relationship between Cr_nmol/kg and {col}')
    plt.show()
```
